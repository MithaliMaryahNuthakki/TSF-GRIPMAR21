# MATCH ANALYSIS

## 1. Number of matches played in each season

```python
ax=sns.countplot(data=match,x='season',palette='summer')
plt.title('Count of Matches in each Season')
plt.figure(figsize=(14,5))
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+0.25))
plt.show()
```

## 2. VENUES

### a)No. of matches at each Venue

```python
plt.figure(figsize=(13,13))
ax0=sns.countplot(data=match,y='venue',palette='winter')
for p in ax0.patches:
    ax0.annotate(format(p.get_width()), (p.get_width()+0.5,p.get_y()+0.5, ))
plt.title('Count of matches played at each venue')
plt.show()
```

### b)Venues with most and least matches

```python
v=match['venue'].value_counts()
print(v[[0,-1]])
```

### c) Number of seasons played in different cities

```python
plt.figure(figsize=(14,6))
ax1=match.groupby('city')['season'].nunique().plot(kind='bar', title ="Number of seasons played in different cities",figsize=(16,5), fontsize=12,color='#B0993C')
for p in ax1.patches:
    ax1.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+0.05))
plt.show()
```

##  3.Team Statistics

```python
team_stats = pd.DataFrame({'Total Matches played': match.team1.value_counts() + match.team2.value_counts(), 'Total won': match.winner.value_counts(), 'Toss won': match.toss_winner.value_counts(), 
                          'Total lost': ((match.team1.value_counts() + match.team2.value_counts()) - match.winner.value_counts())})
team_stats = team_stats.reset_index()
team_stats.rename(columns = {'index':'Teams'}, inplace = True)
team_stats
```

### a) Total Number of matches won by each team

```python
Total_won = match.winner.value_counts()
ax2 = Total_won.plot(kind='bar', title ="No. of wins by each team - Overall",figsize=(10,5), fontsize=12,color='#8DD89D')
ax2.set_xlabel("Teams",fontsize=12)
ax2.set_ylabel("count",fontsize=12)
for p in ax2.patches:
    ax2.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.25))
plt.show()
```

###  b) Total Number of matches lost by each team

```python
Total_lost = ((match.team1.value_counts() + match.team2.value_counts()) - match.winner.value_counts())
ax3 = Total_lost.plot(kind='bar', title ="No. of losses by each team - Overall",figsize=(10,5), fontsize=12,color='#8DD89D')
ax3.set_xlabel("Teams",fontsize=12)
ax3.set_ylabel("count",fontsize=12)
for p in ax3.patches:
    ax3.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.25))
plt.show()
```

### c) Overall winners each season

```python
overall_wins = match.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']].reset_index(drop=True)
overall_wins
```

### d) Number of time a team won the finals

```python
final_matches=match.drop_duplicates(subset=['season'], keep='last')
final_matches["winner"].value_counts()
ax4 = final_matches["winner"].value_counts().plot(kind='bar', title ="Number of Season Wins",figsize=(10,5), fontsize=12,color='#8DD89D')
ax4.set_xlabel("Teams",fontsize=12)
ax4.set_ylabel("count",fontsize=12)
for p in ax4.patches:
    ax4.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.25))
plt.show()
```

## 4. TOSS RESULT 

### a)Toss Selection

```python
Toss=match.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#ff7f0e', '#1f77b4']
plt.figure(figsize = (6,6))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=20)
plt.show()
```

### b)Toss Wins and Winning match

```python
wins = match['toss_winner'] == match['winner']
ax5= wins.value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6),shadow=True)
```

### c) Decision of toss in each season

```python
plt.subplots(figsize=(10,5))
sns.countplot(x="season",hue="toss_decision",data=match)
plt.show()
```

### d)Toss wins by each team

```python
Total_toss_won = match.toss_winner.value_counts()
ax6 = Total_toss_won.plot(kind='bar', title ="No. of toss wins by each team(Overall)",figsize=(10,7), fontsize=12, color='#1f77b4')
ax6.set_xlabel("Teams",fontsize=12)
ax6.set_ylabel("count",fontsize=12)
for p in ax6.patches:
    ax6.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.25))
plt.show()
```

## 7. Batsmen analysis

### a) Top 10 run scorers

```python
ax7= batsman_score=dl.groupby('batsman')['batsman_runs'].agg(['sum']).sort_values('sum',ascending=False).head(10).plot(kind="bar", title="Top 10 run scorers", figsize=(8,5), color="#2e8b7e")

for p in ax7.patches:
    ax7.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2))
plt.show()
```

### b) Players with highest number of 4's

```python
ax9=dl[dl["batsman_runs"] == 4].groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head(10).plot(kind="bar", title="Top 10 players with most 4s", figsize=(8,5), color="#2e8b7e")
for p in ax9.patches:
    ax9.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2))
plt.show()
```

### c) Players with highest number of 6's

```python
ax10=dl[dl["batsman_runs"] == 6].groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head(10).plot(kind="bar", title="Top 10 Players with most 6s", figsize=(8,5), color="#2e8b7e")
for p in ax10.patches:
    ax10.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2))
plt.show()
```

### d) Batsmen Balls faced Vs Runs

```python
batsmen_sum = dl.groupby("batsman").agg({'ball': 'count','batsman_runs': 'sum'})
batsmen_sum.rename(columns={'ball':'balls', 'batsman_runs': 'runs'}, inplace=True)
batsmen_sum = batsmen_sum.sort_values(['balls','runs'], ascending=False)
batsmen_sum.head(10)
```

```python
colors=['#2e8b7e','#ffec8d']
ax11 = batsmen_sum.head().plot(kind='bar', title ="Top 10 Run Scorers (balls Vs runs)",figsize=(16,5), fontsize=12, color=colors)

for p in ax11.patches:
    ax11.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2))
plt.show()
```

### e) Bastman Strike Rate

```python
batsmen_strike = dl.groupby("batsman").agg({'batsman_runs': 'sum'})
batsmen_strike.rename(columns={ 'batsman_runs': 'runs'}, inplace=True)
batsmen_strike['strike_rate'] = batsmen_strike['runs']/batsmen_sum['balls'] * 100
batsmen_strike['strike_rate'] = batsmen_strike['strike_rate'].round(2)
batsmen_strike = batsmen_strike.sort_values(['runs'], ascending=False)
batsmen_strike.head(10)
```

```python
colors=['#ffec8d','#2e8b7e']
ax11 = batsmen_strike.head(10).plot(kind='bar', title ="Top 10 Run Scorers(Runs vs Strike rate)",figsize=(16,5), fontsize=12, color=colors)

for p in ax11.patches:
    ax11.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2))
plt.show()
```

## 8. Bowling Analysis

```python
dismissal_type= dl['dismissal_kind'].value_counts()
ax12 = dismissal_type.head(10).plot(kind='bar', title ="Type Of Dismissals",figsize=(16,5), fontsize=12, color='#DC943D')

for p in ax12.patches:
    ax12.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2))
plt.show()
```

### a) Highest Wicket Takers

```python
wicket_taken=dl[(dl["dismissal_kind"] == 'caught') | (dl["dismissal_kind"] == 'bowled') | (dl["dismissal_kind"] == 'run out') | (dl["dismissal_kind"] == 'lbw') | (dl["dismissal_kind"] == 'stumped') | (dl["dismissal_kind"] == 'caught and bowled') | (dl["dismissal_kind"] == 'hit wicket')].groupby("bowler")['bowler'].count().sort_values(ascending = False)
plt.figure(figsize=(14,8))
ax13 = wicket_taken.head(10).plot(kind='bar', title ="Top 10 Wicket Takers",figsize=(16,5), fontsize=12, color='#DC943D')
for p in ax13.patches:
    ax13.annotate(format(p.get_height()), (p.get_x(), p.get_height()+1))
plt.show()
```
