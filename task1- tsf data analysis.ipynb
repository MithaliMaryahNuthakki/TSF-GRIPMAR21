### Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

%matplotlib inline
```

### Importing The Dataset

```python
data = "http://bit.ly/w-data"
ds = pd.read_csv(data)
```

##### head()- Displays the first 5

```python
ds.head()
```

##### tail() - Displays the last 5 rows

```python
ds.tail()
```

Checking the description of the imported data set

```python
ds.describe()
```

#### Checking for Null Values

```python
ds.isnull().sum()
```

There is no need of cleaning up the dataset, as there are no null values present.

## Visualization of the dataset

#### To find manual relationship between the dataset

```python
plt.scatter(ds['Hours'], ds['Scores'])
plt.title('Hours vs Percentage')
plt.xlabel('Studied Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

#### Regressor plot to determine the relationship between hours studied and scores.

```python
sns.regplot(x="Hours", y="Scores", data=ds);
plt.title('Study Hours vs Percentage Scores')  
plt.grid(True)
plt.show()
```

This plot clearly shows a positive-sloped linear graph was detected between "Hours & Scores", i.e. more hours studied implis more scores.

### Pre Processing the Data

1. Divide the dataset into "attributes" (inputs) and "labels" (outputs
2. Split this data into training and test sets.

```python
X = ds.iloc[:, :-1].values  
y = ds.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0)
```

##### Display the training set 

```python
print(X_train, y_train)
```

##### Display the Test set

```python
print(X_test, y_test)
```

### Training The Model

```python
linearRegressor = LinearRegression()
linearRegressor.fit(X_train, y_train)
y_predict = linearRegressor.predict(X_train)
```

### Training The Algorithm

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete!!!")
```

### Plotting The Regression Line

```python
plt.scatter(X_train, y_train, color="purple")
plt.plot(X_train, regressor.predict(X_train), color = "yellow")
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

### Making Predictions

```python
y_pred = regressor.predict(X_test)
print("Predicted Values: ", y_pred)
```

### Original Values And Predicted Values

#### Tabular Form

```python
oldVsPredicted = pd.DataFrame({"Original Values":y_test,"Predicted Values":y_pred})
print(oldVsPredicted)
```

#### Graphical Form

```python
oldVsPredicted.plot(kind = 'bar', color= ("green","blue"))
plt.grid(True)
```

## Predicting the score for 9.25 hours

```python
print('Score of student who studied for 9.25 hours a day :', regressor.predict([[9.25]]))
```

# Therefore the predicted score is 93.692% for the student who studies for 9.25 hours.

### Evaluation of the model

```python
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:',metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:',
      np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
